## **Normalization**
---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("glass_data.csv")
```

```python

X = df[['Si','Na']]
y = df.iloc[:,[9]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,random_state=0)
```

### **Normalization**
---

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns=X_train.columns)
X_train_scaled.head(5)
```

### **Before Normalization Data Description**
---

```python
np.round(X_train.describe())
```

### **After Normalization Data Description**
---

```python
np.round(X_train_scaled.describe())
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.scatter(X_train['Si'],X_train['Na'],c=y_train.values)
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Si'],X_train_scaled['Na'],c=y_train.values)
ax2.set_title("After Scaling")

plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.set_title("Si Distribution Before Scaling")
sns.kdeplot(X_train['Si'],ax=ax1)
ax2.set_title("Si Distribution After Scaling")
sns.kdeplot(X_train_scaled['Si'],ax=ax2)

plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.set_title("Na Distribution Before Scaling")
sns.kdeplot(X_train['Na'],ax=ax1)
ax2.set_title("Na Distribution After Scaling")
sns.kdeplot(X_train_scaled['Na'],ax=ax2)

plt.show()
```
